```python
import pandas as pd
```

```python
df = pd.read_csv('Gujrat_dataset.csv')
```

```python
df
```

```python
df1 = df.drop(['Unnamed: 0','Date', 'State', 'State_Name'] , axis=1)
```

```python
Confirmed = df['Confirmed']
Deceased = df['Deceased']
Recovered = df['Recovered']
```

```python
csum = Confirmed.sum()
csum
```

```python
dsum = Deceased.sum()
dsum
```

```python
rsum = Recovered.sum()
rsum
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
labels = ['Confirmed','Deceased','Recovered']
sizes = [csum,dsum,rsum]
colors = ['gold', 'yellowgreen', 'lightcoral']
explode = (0.1, 0.1, 0.1) 
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True)
plt.axis('equal')
plt.title('Gujrat State Covid Analysis')
plt.show()
plt.savefig('PICHRT.png')
```
